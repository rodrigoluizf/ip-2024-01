```python
Questão 2
package main

import (
	"bufio"
	"fmt"
	"os"
	"strconv"
	"strings"
)

func main() {
	reader := bufio.NewReader(os.Stdin)
	testCasesInput, _ := reader.ReadString('\n')
	testCases, _ := strconv.Atoi(testCasesInput[:len(testCasesInput)-1])

	for i := 1; i <= testCases; i++ {
		input, _ := reader.ReadString('\n')
		values := strings.Fields(input)
		totalPeople, _ := strconv.Atoi(values[0])
		popular, _ := strconv.ParseFloat(values[1], 64)
		geral, _ := strconv.ParseFloat(values[2], 64)
		arquibancada, _ := strconv.ParseFloat(values[3], 64)
		cadeiras, _ := strconv.ParseFloat(values[4], 64)

		revenue := popular/100*float64(totalPeople)*1 +
			geral/100*float64(totalPeople)*5 +
			arquibancada/100*float64(totalPeople)*10 +
			cadeiras/100*float64(totalPeople)*20

		fmt.Printf("A RENDA DO JOGO N. %d E = %.2f\n", i, revenue)
	}
}
```

```python
Questão 5
package main

import (
	"bufio"
	"fmt"
	"os"
	"strconv"
	"strings"
)

func main() {
	reader := bufio.NewReader(os.Stdin)
	input, _ := reader.ReadString('\n')
	values := strings.Fields(input)
	account, _ := strconv.Atoi(values[0])
	consumption, _ := strconv.ParseFloat(values[1], 64)
	consumerType := values[2]

	var cost float64
	switch consumerType {
	case "R":
		cost = 5 + consumption*0.05
	case "C":
		if consumption <= 80 {
			cost = 500
		} else {
			cost = 500 + (consumption-80)*0.25
		}
	case "I":
		if consumption <= 100 {
			cost = 800
		} else {
			cost = 800 + (consumption-100)*0.04
		}
	}

	fmt.Printf("CONTA = %d\n", account)
	fmt.Printf("VALOR DA CONTA = %.2f\n", cost)
}
```

```python
Questão 8
package main

import (
	"fmt"
	"math"
)

func main() {
	var raio, altura float64
	fmt.Scan(&raio)
	fmt.Scan(&altura)

	const custoAluminioPorM2 = 100.0
	const pi = 3.14159

	areaCirculo := pi * math.Pow(raio, 2)
	areaLateral := 2 * pi * raio * altura
	areaTotal := 2*areaCirculo + areaLateral

	custo := areaTotal * custoAluminioPorM2

	fmt.Printf("O VALOR DO CUSTO E = %.2f\n", custo)
}
```

```python
Questão 19
package main

import (
	"bufio"
	"fmt"
	"os"
	"strconv"
)

func main() {
	reader := bufio.NewReader(os.Stdin)
	input, _ := reader.ReadString('\n')
	n, err := strconv.Atoi(input[:len(input)-1])

	if err != nil || n <= 1 {
		fmt.Println("Numero invalido!")
	} else {
		sum := 0.0
		for i := 1; i <= n; i++ {
			sum += 1 / float64(i)
		}
		fmt.Printf("%.6f\n", sum)
	}
}
```

```python
Questão 20
package main

import (
	"bufio"
	"fmt"
	"os"
	"strconv"
)

func main() {
	reader := bufio.NewReader(os.Stdin)
	hoursInput, _ := reader.ReadString('\n')
	minutesInput, _ := reader.ReadString('\n')
	secondsInput, _ := reader.ReadString('\n')

	hours, _ := strconv.Atoi(hoursInput[:len(hoursInput)-1])
	minutes, _ := strconv.Atoi(minutesInput[:len(minutesInput)-1])
	seconds, _ := strconv.Atoi(secondsInput[:len(secondsInput)-1])

	totalSeconds := hours*3600 + minutes*60 + seconds
	fmt.Printf("O TEMPO EM SEGUNDOS E = %d\n", totalSeconds)
}
```
